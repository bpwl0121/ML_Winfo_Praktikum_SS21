```python
import pandas as pd
import numpy as np
import glob
```

```python
# read the data about all tickets
all_file = pd.read_csv("tickets/tickets.csv")
```

```python
# All tickets
len(all_file)
```

```python
# The "SOLMAN_BTC" user generates a lot of Alert tickets that are not related to the project
# Remove alert tickets generated by the "SOLMAN_BTC" user
all_valid_file = all_file.loc[all_file["Angelegt Von"] != "SOLMAN_BTC "]
```

```python
# Remaining ticekts excluding alert tickets
len(all_valid_file)
```

```python
# Get the file URL of all the tickets
files = glob.glob("tickets/[28]*.csv")
```

```python
# An empty dataframe holds the first 'Beschreibung 'or Nachrichtentyp=='E-Mail von Kunde ' of all the tickets obtained
extracted_df = pd.DataFrame(
    columns=["Nachrichtentyp", "Datun", "Uhrzeit", "Absender", "Ort", "Text", "ID"])


# There are empty tickets and tickets without 'Beschreibung 'or Nachrichtentyp=='E-Mail von Kunde '
# Count empty tickets
num_empty = 0
# Count tickets that do not contain 'Beschreibung ' or Nachrichtentyp=='E-Mail von Kunde '
num_dont_have = 0

# Read non-alert tickets
for file_id in all_valid_file["ID"]:
    try:
        # try to read the first 'Beschreibung 'or Nachrichtentyp=='E-Mail von Kunde ' of each ticket
        df = pd.read_csv("tickets/"+str(file_id)+".csv")

    except Exception as e:
        # the file is empty
        print("this file is empty: "+str(file_id))
        num_empty += 1

    else:
        # get the first 'Beschreibung 'or Nachrichtentyp=='E-Mail von Kunde ' of each ticket
        curr_df = df.query(
            "Nachrichtentyp=='Beschreibung 'or Nachrichtentyp=='E-Mail von Kunde '").head(1)
        curr_df['ID'] = file_id
        # if the tickets without 'Beschreibung 'or Nachrichtentyp=='E-Mail von Kunde '
        if (len(curr_df) == 0):
            print(
                "this file does not have 'Beschreibung 'or Nachrichtentyp=='E-Mail von Kunde ': "+str(file_id))
            num_dont_have += 1
        # Add valid first 'Beschreibung 'or Nachrichtentyp=='E-Mail von Kunde ' of each ticket to the dataframe
        extracted_df = extracted_df.append(curr_df, ignore_index=True)
```

```python
# The size of the saved dataframe, the number of empty files, and the number without 'Beschreibung 'or Nachrichtentyp=='E-Mail von Kunde '.
# The three add up to all the original input all_valid_file
len(extracted_df)+num_empty+num_dont_have == len(all_valid_file)
```
